```python
# Membuat Adjacency matrix
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Membuat urutan node
nodes = list(graph.keys())

# Membuat matrix kosong
matrix = []

for i in nodes:
    row = []
    for j in nodes:
        if j in graph[i]:
            row.append(1)
        else:
            row.append(0)
    matrix.append(row)

# Menampilkan matrix
print("  ", " ".join(nodes))

for i in range(len(nodes)):
    print(nodes[i], matrix[i])
```

```python
graph = {
    'A' : ['B','C'],
    'B' : ['A','D','C'],
    'C' : ['A','E','D','B'],
    'D' : ['B','E','C'],
    'E' : ['D','C']
}
#1. Menampilkan fungsi graph
def print_graph (graf):
  for node in graf:
    print (f" {node} -> {graf [node]}")
print("1. Menampilkan fungsi graph")
print_graph (graph)
from collections import deque
def bfs (graph, start):
  kunjungan = set()
  queue = deque ([start])
  urutan = []
  while queue:
    simpul = queue.popleft ()
    if simpul not in kunjungan:
      print ("List kunjungan:",kunjungan)
      urutan.append(simpul)
      kunjungan.add (simpul)
      queue.extend( [neighbor for neighbor in graph [simpul] if neighbor not in kunjungan])
  return urutan
print("3. Menampilkan BFS dengan mengembalikan nilai, tidak hanya mencetak")
print("Menampilkan BFS:")
hasil= bfs (graph, 'A')
print(hasil)
def find_path(graph, start, end, path=[]):
  # Menambahkan node sekarang ke jalur
  path = path + [start]
  # Jika start sama dengan end
  if start == end:
    return path
  # Jika start tidak ada di graph
  if start not in graph:
    return None

  # Menelusuri tetangga
  for neighbor in graph[start]:
    if neighbor not in path:
      path_baru = find_path(graph, neighbor, end, path)

      # Jika jalur ditemukan
      if path_baru:
        return path_baru
  # Jika tidak ada jalur
  return None
findPath = find_path(graph, 'A', 'E')
print("\n4. Menampilkan fungsi find_path yang mengembalikan satu jalur dari start ke end")
print(findPath)
def is_connected(graph):
   visited = set()
   # Ambil satu simpul awal
   start_node = next(iter(graph))
   # Jalankan DFS
   dfs(graph, start_node, visited)
   # Cek apakah semua simpul sudah dikunjungi
   return len(visited) == len(graph)
print("5. Mengecek apakah semua graph saling terhubung")
print(is_connected(graph))
```
