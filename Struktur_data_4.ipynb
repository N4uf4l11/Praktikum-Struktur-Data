```python
# queue dengan list

# Python program to demonstrate queue implementation using list
queue = []

# append() function to push element in the queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('Initial queue')
print(queue)

# pop() function to pop element fro, queue in LIFO order
print('\nElements poped from queue:')
print(queue.pop())
print(queue.pop())
print(queue.pop())

print('\nqueue after elements are poped:')
print(queue)
# uncommenting print(queue.pop()) will cause an IndexError as the stack is now empty
```

```python
#queue dengan collections.deque

# Python program to demonstrate queue implementation using collections.deque
from collections import deque

queue = deque()

# append() function to push element in the queue
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')

print('Initial queue:')
print(queue)

# pop() function to pop element from queue in LIFO order
print('\nElements poped from queue:')
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())

print('\nqueue after elements are poped:')
print(queue)

# uncomment below to print elements from queue
```

```python
# queue dengan queue.LifoQueue

# Python program to demonstrate queue implementation using queue.LifoQueue
from queue import Queue

# initializing a queue
qu = Queue(maxsize=5)

# qsize() show the number of elements in the queue
print(qu.qsize())


# put() show the number of elements in the queue
qu.put('a')
qu.put('i')
qu.put('u')
qu.put('e')
qu.put('o')


print("Full: ", qu.full())
print("Size: ", qu.qsize())

# get() function to pop element from queue in LIFO order
print('\nElements poped from the queue')
print(qu.get())
print(qu.get())
print(qu.get())

print("\nEmpty: ", qu.empty())
print("Full:",qu.full())
```

```python
class Queue:
  def __init__(self):
    self.queue = list()

  def addtoqu(self,dataval):
    if dataval not in self.queue:
      self.queue.insert(0,dataval)
      return True
    return False

  def removefrom(self):
    if len(self.queue) > 0:
      return self.queue.pop()
    return

  def size(self):
    return len(self.queue)

TheQueue = Queue()
print(TheQueue.addtoqu("Jan"))
print(TheQueue.addtoqu("Feb"))
print(TheQueue.addtoqu("Mar"))
print(TheQueue.addtoqu("Apr"))
print(TheQueue.size())
print(TheQueue.removefrom())
print(TheQueue.removefrom())
print(TheQueue.size())
```
