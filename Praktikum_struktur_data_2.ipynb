```python
# Latihan
# 1. Membuat double linked list

# Membuat linked list
class Node:
  def __init__(self, isi=None):
    self.isi =isi
    self.next_ = None
    self.prev_ = None

class doubleLinkedList:
  def __init__(self):
    self.head = None
  def listprint(self):
    printval = self.head
    while printval is not None:
      print (printval.isi)
      printval = printval.next_

# Penyiipan tengah
  def SisipTengah(self,mid_node,newdata):
    if mid_node is None:
      print("Data tidak tersedia")
      return
    NewNode = Node(newdata)
    NewNode.next_ = mid_node.next_
    NewNode.prev_ = mid_node
    if mid_node.next_ is not None:
      mid_node.next_.prev_ = NewNode
    mid_node.next_ = NewNode

# Hapus sesuai Kunci
  def RemoveNode(self, Removekey):
    current = self.head

    if current is not None and current.isi == Removekey:
        self.head = current.next_
        if self.head is not None:
            self.head.prev_ = None
        current = None
        return
    while current is not None and current.isi != Removekey:
        current = current.next_
    if current is None:
        return
    if current.prev_ is not None:
        current.prev_.next_ = current.next_
    if current.next_ is not None:
        current.next_.prev_ = current.prev_

    current = None


n1= Node("Sekarang")
n2= Node("Hari ini")
n3= Node("Kemaren")

Li = doubleLinkedList()
Li.head = n1

n1.next_ = n2
n2.prev_ = n1
n2.next_ = n3
n3.prev_ = n2

print("Sebelum dilakukan Sisipan Tengah")
Li.listprint()
print("="*30)
Li.SisipTengah(n1,"Besok")
print("Setelah dilakukan Sisipan Tengah")
Li.listprint()
print("Sebelum dilakukan penghapusan")
print("="*30)
Li.RemoveNode("Hari ini")
print("Setelah dilakukan penghapusan")
Li.listprint()


```
