```
class Node:
  def __init__(self, data):
    self.data = data #pendefinisian node utama
    self.left = None #pendefinisian node sebelah kiri
    self.right = None #pendefinisian node kanan

class BinarySearchTree:
  def __init__(self):
    self.root = None #pendefinisian node pertama untuk pencarian (node induk)
  def insert(self, root, data):
    if root is None:
      return Node(data) #pengembalian data apabila node kosong
    if data < root.data:
      root.left = self.insert(root.left, data) #mengambil node kiri apabila lebih kecil dari node induk
    else:
      root.right = self.insert(root.right, data) #mengambil node kanan apabila lebih besar dari node induk
    return root #pengembalian nilai root untuk ditampilkan

bst = BinarySearchTree() #menyimpan pemanggilan fungsi dalam sebuah variabel untuk memudahkan pemanggilan
root = None #pendefinisian node pertama yang diawali dengan sebuah node kosong
data_list = [40, 10, 10, 70, 65, 12, 7]
for value in data_list:
  root = bst.insert(root, value) #pemanggilan fungsi bst dengan menyimpan didalam variabel
def inorder(node):
  if node:
    inorder(node.left) #mengambil node kiri
    print(node.data, end=' ') #mengambil node induk
    inorder(node.right) #mengambil node kanan
def preorder(node):
  if node:
    print(node.data, end=' ') #mengambil node induk
    preorder(node.left) #mengambil node kiri
    preorder(node.right) #mengambil node kanan
def postorder(node):
  if node:
    postorder(node.left) #mengambil node kiri
    postorder(node.right) #mengambil node kanan
    print(node.data, end=' ') #mengambil node induk
print("Hasil inorder traversal:")
inorder(root) #pemanggilan traversal fungsi inorder
print("\nPreorder Traversal:")
preorder(root) #pemanggilan traversal fungsi preorder
print("\nPostorder Traversal:")
postorder(root) #pemanggilan traversal fungsi postorder
print("\n","="*40)
print("Pencarian key:")
def search(node, key):
  if node is None or node.data == key:
    return node #mengembalikan nilai apabila nilai sesuai key
  if key < node.data:
    return search(node.left, key) #mencari node kiri apabila lebih kecil dari key
    return search(node.right, key) #mencari node kanan apabila lebih kecil dari key
# Uji pencarian
key = 40
result = search(root, key) #pemanggilan fungsi search dan menyimpan dalam variabel
if result:
  print(f"{key} ditemukan dalam tree.") #ditemukan menampilan key ditemukan
else:
  print(f"{key} tidak ditemukan.") #tidak ditemukan menampilkan key tidak ditemukan
key = 45
result = search(root, key) #pemanggilan fungsi search dan menyimpan dalam variabel
if result:
  print(f"{key} ditemukan dalam tree.") #ditemukan menampilan key ditemukan
else:
  print(f"{key} tidak ditemukan.") #tidak ditemukan menampilkan key tidak ditemukan
```
